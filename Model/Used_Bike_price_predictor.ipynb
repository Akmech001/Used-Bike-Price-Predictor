# Model

```python
X=bike1.drop(columns='price')
y=bike1['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe=OneHotEncoder()
ohe.fit(X[['bike_name','brand','city','owner']])

column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['bike_name','brand','city','owner']),remainder='passthrough')
```

```python
lr=LinearRegression()
```

```python

pipe=make_pipeline(column_trans,lr)

pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
import pickle
```

```python
pickle.dump(pipe,open('LinearRegressionModel.pkl','wb'))
```

```python
pipe.predict(pd.DataFrame([['Royal Enfield Himalayan','Royal Enfield','First Owner','441','16000','Chennai','2']],columns=['bike_name','brand','owner','power','kms_driven','city','age']))
```

```python
def predict_price(bike_name, brand, owner, power, kms_driven, city, age):
    # Create a DataFrame with the input data
    data = pd.DataFrame([[bike_name, brand, owner, power, kms_driven, city, age]],
                        columns=['bike_name', 'brand', 'owner', 'power', 'kms_driven', 'city', 'age'])
    # Predict the price
    predicted_price = pipe.predict(data)
    return predicted_price[0]
```
